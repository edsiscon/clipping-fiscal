```python
import requests
import bs4
import pandas as pd

mig = "https://www.migalhas.com.br/quentes"
conplan = "https://coplancontabilidade.com.br/blog/"
contabeis = "https://www.contabeis.com.br/noticias/"
sinfrej = "https://sinfrerj.com.br/destaque-na-imprensa"
totvs = "https://www.totvs.com/blog/fiscal-clientes/"
alerj ="https://www.alerj.rj.gov.br/Comunicacao/SalaDeImprensa/Noticias"

lista_sites = [mig, conplan, contabeis, sinfrej, totvs,alerj]

lista_palavras_chave = ['DFe','SPED','EFD','FECP','FOT','RECEITA FEDERAL','ICMS', 'ICMS-ST','PIS','COFINS','IMPOSTO','TRIBUTO','TRIBUTAÇÃO','FAZENDA','IMPOSTO','IMPOSTOS','NFC','NFC-E','NFE','NOTA TÉCNICA','IBS','CBS','REFORMA TRIBUTÁRIA','RTC']
lista  =[]
lista_texto = []

for item in lista_sites:
  requisicao = requests.get(item)
  pagina = bs4.BeautifulSoup(requisicao.text, "html.parser")

  match lista_sites.index(item):
    case 0:
      lista = pagina.find_all("h2", class_="topico__header")
    case 1:
      lista = pagina.find_all("h4", class_="dlab-title")
    case 2:
      lista = pagina.find_all("div", class_="texto")
    case 3:
      lista = pagina.find_all("h5", class_="card-title mt-2")
    case 4:
      lista = pagina.find_all("h3", class_="title-list-new")
    case 5:
      lista = pagina.find_all("div", class_="subtitulo")

  for link in lista:
    titulo = link.get_text()
    if any(palavra.lower() in titulo.lower() for palavra in lista_palavras_chave):
      lista_texto.append(titulo)
df = pd.DataFrame(lista_texto)
resultado_json = df.to_json()
df.columns = ['Título']
pd.options.display.max_rows = None
pd.options.display.max_columns = None

display(df)

```
